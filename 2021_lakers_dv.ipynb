```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/JBoundless/2021-lakers-dv/main/2021%20Lakers%20v2%20-%20Sheet1.csv')

playerName = df[df.playerName == 'playerName']
fgPer = df[df.fgPer == 'fgPer']
threePer = df[df.threePer == 'threePer']
ftPer = df[df.ftPer == 'ftPer']
pointsPerGame = df[df.PTS == 'PTS']
reboundsPerGame = df[df.TRB == 'TRB']
assistsPerGame = df[df.AST == 'AST']
stealsPerGame = df[df.STL == 'STL']
blocksPerGame = df[df.BLK == 'BLK']
turnoversPerGame = df[df.TOV == 'TOV']

y1 = df.PTS
y2 = df.TRB
y3 = df.AST

plt.bar(df.playerName, y1, color="gold", label="PTS")
plt.bar(df.playerName, y2, color="purple", label="TRB")
plt.bar(df.playerName, y3, color="blue", label="AST")
plt.title('2020-21 Lakers Stats')
plt.xticks(rotation = 'vertical')
plt.legend()
plt.show()
```
